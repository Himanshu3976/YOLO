```
f = open("GroundTruth.txt")
read = f.read()
x = f.read().split(",")
f.close()
lines = read.splitlines()

data_values = []
file_names = []

for l in range(1,len(lines)):
  box = lines[l].split(',')
  box_val = ' '.join(box[1:])
  data_values.append(box_val)
  file_name = "Image" + str(l)+ ".txt"
  file_names.append(file_name)
  
```

```
for i in range(len(file_names)):
  f = open(file_names[i], 'w+')
  f.write("0\n")
  f.write(data_values[i])
  f.close()
 
  
```
